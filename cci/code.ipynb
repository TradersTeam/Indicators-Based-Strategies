```python
from finta import TA
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
btc = yf.Ticker('BTC-USD')
hist = btc.history(period="3mo",interval="1h")
hist
```

```python
cci=TA.CCI(hist,14)
cci
```

```python
hist['cci']=cci
hist
```

```python
def signal(cci_past, cci_current):
    if cci_past < -150 and cci_current > -150:
        return 'buy'
    elif cci_past > 150 and cci_current < 150:
        return 'sell'
    else:
        return 'no signal'
```

```python
hist['idx']=range(0,len(hist))
hist
```

```python
hist['signal'] = hist.apply(lambda row: signal(hist.iloc[int(row['idx']-1)]['cci'],row['cci']), axis=1)
hist
```

```python
def status(signal,status_past):
    if signal=='buy' and (status_past=='cash' or status_past=='sell'):
        return 'buy'
    elif signal=='sell' and (status_past=='hold' or status_past=='buy'):
        return 'sell'
    elif signal=='no signal' and (status_past=='buy' or status_past=='hold'):
        return 'hold'
    elif status_past=='hold':
        return 'hold'
    elif status_past=='cash':
        return 'cash'
    else:
        return 'cash'
```

```python
status_list = []
status_list.append('cash')
for i in range(1, len(hist)-1):
    status_list.append(status(hist.iloc[i]['signal'], status_list[i-1]))
status_list.append('sell')

hist['status'] = status_list

hist
```

```python
bb_list = []
cb_list = []
cb_list.append(float(1000))
bb_list.append(float(0))
for i in range(1, len(hist)):
    status = hist.status[i]
    price = float(hist.Close[i])
    if status == 'buy':
        bb_list.append(float(cb_list[i-1]/price))
        cb_list.append(float(0))
    elif status == 'sell':
        cb_list.append(float(bb_list[i-1]*price))
        bb_list.append(float(0))
    elif status=='cash':
        bb_list.append(float(0))
        cb_list.append(cb_list[i-1])
    elif status=='hold':
        bb_list.append(bb_list[i-1])
        cb_list.append(float(0))
        


hist['bb'] = bb_list
hist['cb'] = cb_list

hist
```
