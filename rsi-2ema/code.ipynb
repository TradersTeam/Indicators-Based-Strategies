```python
# imports
from finta import TA
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
# catch dataframe
btc = yf.Ticker("BTC-USD")
hist = btc.history(period="1y",interval="1d")
hist
```

```python
rsi14=TA.RSI(hist,14)
rsi14
```

```python
ema50=TA.EMA(hist,50)
ema50
```

```python
ema120=TA.EMA(hist,120)
ema120
```

```python
hist['rsi14']=rsi14
hist['ema50']=ema50
hist['ema120']=ema120
hist
```

```python
def trend(a, b):
    if a > b:
        return 'up'
    elif a < b:
        return 'down'
    else:
        return '-'


hist['trend'] = hist.apply(lambda row: trend(row['ema50'], row['ema120']), axis=1)
hist
```

```python
def state(rsi):
    if 0 < rsi < 30:
        return 'oversold'
    elif 70 < rsi < 100:
        return 'overbought'
    else: return '-'
def signal(trend,state):
    if trend=='up' and state=='oversold':
        return 'buy'
    if trend=='down' and state=='overbought':
        return 'sell'
    else:
        return '-'
hist['state']=hist.apply(lambda row: state(row['rsi14']),axis=1)
hist['signal']=hist.apply(lambda row: signal(row['trend'],row['state']),axis=1)
hist
```
