```python
from finta import TA
import numpy as np
import pandas as pd
```

```python
import yfinance as yf
```

```python
btc = yf.Ticker('BTC-USD')
hist = btc.history(period="1mo",interval="30m")
hist
```

```python
rsi=TA.RSI(hist)
rsi
```

```python
stochd=TA.STOCHD(hist,3,11)
stochd
```

```python
hist['rsi']=rsi
hist['stoch']=stochd
hist
```

```python
def signal(rsi_past, rsi_current, stoch):
    if rsi_past < 30 and rsi_current > 30 and stoch > 50:
        return 'buy'
    elif rsi_past > 70 and rsi_current < 70 and stoch < 50:
        return 'sell'
    else:
        return '-'
```

```python
hist['index']=range(0,len(hist))
hist
```

```python
hist['signal'] = hist.apply(lambda row: signal(hist.iloc[int(row['index']-1)]['rsi'],row['rsi'],row['stoch']), axis=1)
hist
```
